```python
import pandas as pd
df = pd.read_csv('data/fighters.csv')
```

```python
df
```

```python
plt.figure(figsize=(6,5))
plt.scatter(df['attack_rate'], df['block_rate'], s=60)
plt.xlabel('Attack Rate')
plt.ylabel('Block Rate')
plt.title('K-Means Clustering: Fighter Play Styles')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df)
```

```python
import matplotlib.pyplot as plt

# Visualize (use first two features for clarity)
plt.figure(figsize=(6,5))
plt.scatter(df['attack_rate'], df['block_rate'], c=df['cluster'], cmap='rainbow', s=60)
plt.xlabel('Attack Rate')
plt.ylabel('Block Rate')
plt.title('K-Means Clustering: Fighter Play Styles')
plt.show()
```

```python
# Print centroids for explanation
print("Cluster Centers:\n", pd.DataFrame(kmeans.cluster_centers_, columns=['attack_rate', 'block_rate', 'move_speed']))
```
